```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
# import data
df = pd.read_csv('/content/zomato.csv', on_bad_lines='skip', engine='python', encoding='latin1')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

```python
# removing duplicates
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#Filling Null Values in Rate Column with Mean
df['rate'] = df['rate'].fillna(df['rate'].mean())
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
# Dropping Null Values
df.dropna(inplace = True)
df.head()
```

```python
# renaming the column
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
# drop listed in city as both have same value and doing same work
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df['Cost2plates'].unique()
```

```python
# removing comma from cost2plates

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '') # Change from ' ' to ''
        return float(value)
    else:
        return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
# cleaning rest type
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
# cleaning rest type
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
# if data less than 1000 in rest type, sum them all and count as others
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
#cleaning location
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
# cleaning cuisine
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
# NLP : natural language process
import pandas as pd
import numpy as np
from textblob import TextBlob
import plotly.graph_objects as go
import plotly.express as xp
```

```python
modi["Tweet"] = modi["Tweet"].astype(str)
rahul["Tweet"] = rahul["Tweet"].astype(str)

def find_polarity(review):
    return TextBlob(review).sentiment.polarity

modi["polarity"] = modi["Tweet"].apply(find_polarity)
rahul["polarity"] = rahul["Tweet"].apply(find_polarity)
modi.head()
```

```python
modi.head()
```

```python
rahul.head()
```

```python
rahul
```

```python
modi['Lable'] = np.where(modi['polarity'] > 0, 'Positive', np.where(modi['polarity'] < 0, 'Negative', 'Neutral'))
rahul['Lable'] = np.where(rahul['polarity'] > 0, 'Positive', np.where(rahul['polarity'] < 0, 'Negative', 'Neutral'))
modi.head()
```

```python
rahul
```

```python
neutral_modi = modi[modi['polarity']==0.000]
remove_neutral_modi = modi['polarity'] . isin(neutral_modi['polarity'])
modi.drop(modi[remove_neutral_modi].index , inplace = True)
```

```python
print(neutral_modi.shape)
print(modi.shape)
```

```python
neutral_rahul = rahul[rahul['polarity']==0.000]
remove_neutral_rahul = rahul['polarity'] . isin(neutral_modi['polarity'])
rahul.drop(rahul[remove_neutral_rahul].index , inplace = True)
```

```python
print(neutral_rahul.shape)
print(rahul.shape)
```

```python
#making row count to 9000 for#modi
np.random.seed(10)
remove_n = modi.shape[0] - 9000
drop_indices = np.random.choice(modi.index, remove_n, replace=False)
df_modi = modi.drop(drop_indices)
```

```python
print(df_modi.shape)
```

```python
#rahul
np.random.seed(10)
remove_n =367
drop_indices = np.random.choice(rahul.index, remove_n, replace=False)
df_rahul = rahul.drop(drop_indices)
```

```python
print(df_rahul.shape)
```

```python
modi_count = df_modi.groupby('Lable').count()
modi_count
```

```python
neg_modi = (modi_count.loc['Negative', 'polarity'] / df_modi.shape[0]) * 100
pos_modi = (modi_count.loc['Positive', 'polarity'] / df_modi.shape[0]) * 100
```

```python
rahul_count = df_rahul.groupby('Lable').count()
rahul_count
```

```python
neg_rahul = (rahul_count.loc['Negative', 'polarity'] / df_rahul.shape[0]) * 100
pos_rahul = (rahul_count.loc['Positive', 'polarity'] / df_rahul.shape[0]) * 100
```
