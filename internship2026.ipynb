```python
# checking the distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

```python
# encoding 'Fuel_Type' column
fuel_type_mapping = {'Petrol':0, 'Diesel':1, 'CNG':2}
car_dataset['Fuel_Type'] = car_dataset['Fuel_Type'].map(fuel_type_mapping).fillna(-1).astype(int)

# encoding 'Seller_Type' column
seller_type_mapping = {'Dealer':0, 'Individual':1}
car_dataset['Seller_Type'] = car_dataset['Seller_Type'].map(seller_type_mapping).fillna(-1).astype(int)

# encoding 'Transmission' column
transmission_mapping = {'Manual':0, 'Automatic':1}
car_dataset['Transmission'] = car_dataset['Transmission'].map(transmission_mapping).fillna(-1).astype(int)
```
