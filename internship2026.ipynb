```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
# import data
df = pd.read_csv('/content/zomato.csv', on_bad_lines='skip', engine='python', encoding='latin1')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

```python
# removing duplicates
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#Filling Null Values in Rate Column with Mean
df['rate'] = df['rate'].fillna(df['rate'].mean())
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
# Dropping Null Values
df.dropna(inplace = True)
df.head()
```

```python
# renaming the column
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
# drop listed in city as both have same value and doing same work
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df['Cost2plates'].unique()
```

```python
# removing comma from cost2plates

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '') # Change from ' ' to ''
        return float(value)
    else:
        return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
# cleaning rest type
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
# cleaning rest type
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
# if data less than 1000 in rest type, sum them all and count as others
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
#cleaning location
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
# cleaning cuisine
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
# loading the data from csv file to pandas dataframe
car_dataset = pd.read_csv('/content/car data.csv')
```

```python
# checking the distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

```python
# encoding 'Fuel_Type' column
fuel_type_mapping = {'Petrol':0, 'Diesel':1, 'CNG':2}
car_dataset['Fuel_Type'] = car_dataset['Fuel_Type'].map(fuel_type_mapping).fillna(-1).astype(int)

# encoding 'Seller_Type' column
seller_type_mapping = {'Dealer':0, 'Individual':1}
car_dataset['Seller_Type'] = car_dataset['Seller_Type'].map(seller_type_mapping).fillna(-1).astype(int)

# encoding 'Transmission' column
transmission_mapping = {'Manual':0, 'Automatic':1}
car_dataset['Transmission'] = car_dataset['Transmission'].map(transmission_mapping).fillna(-1).astype(int)
```

```python
X = car_dataset.drop(['Car_Name' , 'Selling_Price',] ,axis=1)
Y = car_dataset['Selling_Price']
```

```python
#loading the linear regression model
lin_reg_model = LinearRegression()
```

```python
# Force-encode any remaining text strings in the dataset
car_dataset.replace({'Seller_Type': {'Dealer': 0, 'Individual': 1}}, inplace=True)

# Re-split the clean data so X_train updates in memory
X = car_dataset.drop(['Car_Name', 'Selling_Price'], axis=1)
Y = car_dataset['Selling_Price']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2)

# Now train the model
lin_reg_model.fit(X_train, Y_train)
```

```python
# predicting on trainig data
lin_reg_model.fit(X_train, Y_train)
```

```python
# r squared error
training_data_prediction = lin_reg_model.predict(X_train)
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
